```python
file = open('data.json', 'r')
data = json.load(file)
```

```python
data
```

```python
final_set=[]

for item in data:
    for results in data[item]:
        
        
        row={}
        for food_info in results:
            
            if(food_info!='servings'):
                row[food_info]=results[food_info]
                
        if type(results['servings']['serving'])==list:
            for each_serving in results['servings']['serving']:
                row1=row
                for nutri in each_serving:
                    row1[nutri]=each_serving[nutri]
                #print(row1)
                final_set.append(row1.copy())
                
                
        else :
            row1=row
            for nutri in results['servings']['serving']:
                    row1[nutri]=results['servings']['serving'][nutri]
            #print(row1,"\n")
            final_set.append(row1)
```

```python
final_set
```

```python
pd.DataFrame.from_dict(final_set, orient='columns').to_csv('trail.csv')
```
